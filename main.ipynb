```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
import ollama
```

```python
MODEL = "llama3"
```

```python
class Website:
    url: str
    title: str
    text: str

    def __init__(self, url):
        self.url = url
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.text = soup.body.get_text(separator="\n", strip=True)
```

```python
cnn = Website("https://cnn.com")
print(cnn.title)
print(cnn.text)
```

```python
system_prompt = "You are a covert CIA operative intercepting web data. \
                Extract the core intel from the provided website content and generate a rapid Situation Report (SITREP) summary. \
                Disregard operational decoys like navigation links and boilerplate text. \
                Format your response as a classified markdown dossier."
```

```python
def user_prompt(website):
    user_prompt=f"You are looking at a website with title {website.title} "
    user_prompt += "and the content of this website is as follows:\
                 please provide a short summary of this website in markdown.\n\n"
    user_prompt +=website.text
    return user_prompt
    
```

```python
def messages_for(website):
    return[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt(website)}
    ]
```

```python
def summarize(url):
    website = Website(url)
    messages = messages_for(website)
    response= ollama.chat(MODEL, messages=messages)
    return response['message']['content']
```

```python
summarize("https://cnn.com")
```

```python
def display_summary(url) :
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://cnn.com")
```
